## Exploring RFM Analysis (Recency, Frequency, Monetary Value)

````python
# Assuming `sales_data` is your DataFrame and `TotalSales` column represents the revenue for each transaction.

# Define the date for calculating recency (e.g., last transaction date in the dataset)
current_date = sales_data['InvoiceDate'].max()

# Calculate RFM metrics
rfm = sales_data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (current_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'TotalSales': 'sum'  # Monetary Value
}).reset_index()

# Rename columns
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Display the RFM table
print(rfm.head())
````

````python
# Define RFM scoring with equal-width bins
rfm['R_Score'] = pd.cut(rfm['Recency'], bins=5, labels=[5, 4, 3, 2, 1])  # Recency
rfm['F_Score'] = pd.cut(rfm['Frequency'], bins=5, labels=[1, 2, 3, 4, 5])  # Frequency
rfm['M_Score'] = pd.cut(rfm['Monetary'], bins=5, labels=[1, 2, 3, 4, 5])  # Monetary

# Combine scores to create the RFM score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Display the updated RFM table
print(rfm.head())
````
